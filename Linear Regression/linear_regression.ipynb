```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LinearRegression
import pickle
from google.colab import files

# Load dataset
data = fetch_california_housing()
X = data.data
y = data.target

# ---------------- SIMPLE LINEAR REGRESSION ----------------
X_simple = X[:, 0].reshape(-1, 1)

# Manual OLS
X_b = np.c_[np.ones((len(X_simple), 1)), X_simple]
theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

y_pred_manual = X_b.dot(theta)

# Sklearn
model = LinearRegression()
model.fit(X_simple, y)
y_pred_sklearn = model.predict(X_simple)

# Plot
plt.scatter(X_simple, y)
plt.plot(X_simple, y_pred_sklearn)
plt.title("Simple Linear Regression")
plt.show()

# ---------------- MULTIPLE LINEAR REGRESSION ----------------
model_full = LinearRegression()
model_full.fit(X, y)

# Print comparison
print("Manual Theta:", theta)
print("Sklearn Coef:", model.coef_, model.intercept_)

# Save model
with open("model.pkl", "wb") as f:
    pickle.dump(model_full, f)

print("Model saved as model.pkl")

# 🔥 AUTO DOWNLOAD
files.download("model.pkl")
```
